```python
import tweepy
from twitter_authentication import bearer_token
import time
import pandas as pd
import openpyxl
import numpy as np
```

```python
client = tweepy.Client(bearer_token, wait_on_rate_limit=True)
```

```python
ankarabbld_Tweets = []
for response in tweepy.Paginator(client.search_all_tweets, 
                                 query = 'to:434788572 -is:retweet',                                 
                                 user_fields = ['username', 'public_metrics'],
                                 tweet_fields = ['created_at', 'public_metrics', 'text', 'conversation_id'],
                                 expansions = ['author_id', 'in_reply_to_user_id'],
                                 start_time = '2020-05-01T00:00:00Z',
                                 end_time = '2020-05-02T00:00:00Z',
                              max_results=500):
    time.sleep(1)
    ankarabbld_Tweets.append(response)
```

```python
ankarabbld_Tweets[0].data[0]
```

```python
result = []
user_dict = {}
for response in ankarabbld_Tweets:
    for user in response.includes['users']:
        user_dict[user.id] = {'username': user.username, 
                              'followers': user.public_metrics['followers_count'],
                              'tweets': user.public_metrics['tweet_count'],                              
                             }
    for tweet in response.data:
        author_info = user_dict[tweet.author_id]       
        result.append({'author_id': str(tweet.author_id),
                       'author_info_username': author_info['username'],
                       'tweet_id':str(tweet.id),
                       'tweet_text': tweet.text,
                       'tweet_conversation_id': str(tweet.conversation_id),                       
                       'in_repy_to_user_id':str(tweet.in_reply_to_user_id),                                        
                       'tweet_created_at': tweet.created_at.strftime("%d/%m/%Y, %H:%M:%S"),                                                       
                       'retweet_count': tweet.public_metrics['retweet_count'],
                       'reply_count': tweet.public_metrics['reply_count'],
                       'like_count': tweet.public_metrics['like_count'],
                       'quote_count': tweet.public_metrics['quote_count']
                       
                      })
        
    
df = pd.DataFrame(result)
```
